```python
note_roots = {
    'C0': 0,
    'C#': 1,
    'Db': 1,
    'D0': 2,
    'D#': 3,
    'Eb': 3,
    'E0': 4,
    'E#': 5,
    'Fb': 4,
    'F0': 5,
    'F#': 6,
    'Gb': 6,
    'G0': 7,
    'G#': 8,
    'Ab': 8,
    'A0': 9,
    'A#': 10,
    'Bb': 10,
    'B0': 11,
    'B#': 0,
    'Cb': 11
}

keys = {
    0: (0, 21),
    1: (7, 16),
    2: (2, 23),
    3: (9, 18),
    4: (4, 13),
    5: (11, 20),
    6: (6, 15),
    7: (1, 22),
    -1: (5, 14),
    -2: (10, 19),
    -3: (3, 12),
    -4: (8, 17),
    -5: (1, 22),
    -6: (6, 15),
    -7: (11, 20)
}

noteroots = np.array(['C0', 'Db', 'D0', 'Eb', 'E0', 'F0', 'F#', 'G0', 'Ab', 'A0', 'Bb', 'B0'])
```

```python
def data_for_file(path):
  df = pd.read_csv(path).to_numpy()

  df_dict_notes = {}
  df_dict_chords = {}

  max_measure_notes = 0

  for i in df:

    def find_key(root, label):
      if label == 'minor':
        return keys[root][1]
      else:
        return keys[root][0]
                 

    key = find_key(i[2], i[3])

    key_root = key % 12
  
    if df_dict_notes.get(i[1]) == None:
      df_dict_notes[i[1]] = []
    if i[6] in note_roots.keys():
      note = note_roots[i[6]]

      # normalize to key position
      note -= key_root
      if note < 0:
        note += 12

      df_dict_notes[i[1]].append(note)

    if df_dict_chords.get(i[1]) == None:
      if i[4] != '[]':
        if i[5] != '[]':
          chord = note_roots[i[4]]

          # normalize to key position
          chord -= key_root
          if chord < 0:
            chord += 12
          
          if i[5] == 'minor':
            chord += 12

          df_dict_chords[i[1]] = chord


    max_measure_notes = max(max_measure_notes, len(df_dict_notes[i[1]]))

  labels = []
  values = []

  for i in list(df_dict_chords.keys()):
    labels.append(df_dict_chords[i])
    values.append(df_dict_notes[i])
    values[-1] = np.pad(values[-1], (0 ,max_measure_notes - len(values[-1])), 'constant', constant_values=-1)
  labels = np.array(labels).astype(int)
  values = np.array(values).astype(int)

  print("parsed", path)
    
  return values, labels
```

```python
def data_for_path(folder, count):
    path = root_dir + folder 
    x = []
    y = []
    counter = 0
    for _, _, files in os.walk(path):
        for file in files:
            counter += 1
            if count > -1:
                if counter > count:
                    continue
            print(counter)
            v, l = data_for_file(path+file)
            for i in l:
              y.append(i)
            for j in v:
                x.append(j)
    return x, y
```

```python
x_train, y_train = data_for_path('/csv_train/', 1000)
x_test, y_test = data_for_path('/csv_test/', 100)
```

```python
def length_N(x, y, N):
    x_out = []
    y_out = []
    for i in range(len(x)):
        sequence = x[i]
        for j in range(len(sequence)-(N-1)):
            if sequence[j+(N-1)] != -1:
                arr = []
                for k in range(N):
                    arr.append(sequence[j+k])
                x_out.append(arr)
                y_out.append(y[i])
    return np.array(x_out), np.array(y_out)
```

```python
x_train, y_train = length_N(x_train, y_train, 4)
x_test, y_test = length_N(x_test, y_test, 4)

print(x_train.shape)
```

```python
x_save = x_train
y_save = y_train
```

```python
x_train = x_save
y_train = y_save
```

```python
def shape_input(values, labels):

#   for i in range(len(values)):
#     values[i] = np.pad(values[i], (0 ,max_seq_len - values[i].size), 'constant', constant_values=-1)

  labels = np.array(labels)
  values = np.array(values)

  network_input = np.reshape(values, (values.shape[0], values.shape[1], 1))
  network_output = np.zeros((labels.shape[0], 24))

  for i in range(labels.shape[0]):
    network_output[i][labels[i]] = 1    
    
  return network_input, network_output
```

```python
x_train, y_train = shape_input(x_train, y_train)

x_test, y_test = shape_input(x_test, y_test)
```

```python
from keras.models import Sequential
from keras.layers import TimeDistributed
from keras.layers import Bidirectional
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import LSTM
from keras.layers import Flatten
from keras.layers import Activation
from keras.utils import np_utils
from keras.callbacks import ModelCheckpoint
from keras import optimizers
```

```python
model = Sequential()
model.add(Bidirectional(LSTM(512,input_shape=(x_train.shape[1], x_train.shape[2]),return_sequences=True, activation='tanh')))
model.add(Dropout(0.4))
model.add(Bidirectional(LSTM(256, return_sequences=False, activation='tanh')))
model.add(Dropout(0.4))
model.add(Dense(24, activation='softmax'))
# adam = optimizers.Adam(learning_rate=0.01, beta_1=0.9, beta_2=0.999, amsgrad=False)
model.compile(loss='categorical_crossentropy', optimizer= 'adam', metrics = ['accuracy'])


history = model.fit(x=x_train, y=y_train, validation_data=(x_test, y_test), epochs=15, batch_size=64) #,  callbacks=callbacks_list)
model.save("model_15epochs.h5")
```
